```python
import argparse, json, logging, csv, re, sys, codecs

floatre = re.compile("^\d+\.\d+$")
intre = re.compile("^\d+$")

def read_header(file="h.txt"):
    header=[]
    for line in open(file):
        header.append(line.strip())
    logging.info("%d lines in header", len(header))
    return header

def process_csv(file, header):
    out=[]
    stdin = file == "-"
    fd = sys.stdin if stdin else codecs.open(file, 'r', 'UTF-8')
    reader = csv.reader(fd)
    for nr, row in enumerate(reader):
        logging.debug("%d fields in line %d", len(row), nr)
        d = dict()
        out.append(d)
        for i, field in enumerate(row):
            if field != "NULL":
                if floatre.match(field):
                    d[header[i]] = float(field)
                elif intre.match(field):
                    d[header[i]] = int(field)
                else:
                    d[header[i]] = field
    if not stdin:
        fd.close()
    return out

'''
if __name__ == "__main__":
    p =argparse.ArgumentParser()
    p.add_argument("raw", nargs="*", default=["-"])
    p.add_argument("--header", type=str, default="h.txt")
    p.add_argument("--debug", action="store_true")
    args = p.parse_args()
    if args.debug:
        logging.basicConfig(level=logging.INFO)
    header = read_header(args.header)
    out = []
    for file in args.raw:
        out += process_csv(file, header)
    print(json.dumps(out, indent=4, ensure_ascii=True))
''' 

header = read_header('h.txt')
data = process_csv('years/1964', header)
for element in data:
    if 'award_label' in element:
        #print(element['award_label'])
        if "academy award" in element['award_label'].lower():
            print(element)
    if 'Crowe' in element['rdf-schema#label']:
-        print(element)
```

```python
import json
import csv 

with open ('nobel.json') as file:
    nobel_data= json.load(file)

nobel_dict = {}
for prize in nobel_data['prizes']:
    for laureate in prize['laureates']:
        nobel_dict[laureate['firstname'].split(' ')[0] + ' ' + laureate['surname']] = prize

print(nobel_dict['David Thouless']['year'])
```

```python
header = read_header('h.txt')
nobel_sorted = []
for year in range(1820, 2014):
    print(year)
    
    data = process_csv('years/' + str(year), header)
    for elements in data:
        names = elements['rdf-schema#label'].split (' ') 
        name= names[0] + ' ' + names[-1] 
        gender = 'F' if 'WomenNobelLaureates' in elements['22-rdf-syntax-ns#type_label'] else 'M'
        birthYear = elements['birthDate'].split('-')[0][-4:]
        has_label = 'award_label' in elements and "nobel" in elements['award_label'].lower()
        has_ns_label = "nobel" in elements['22-rdf-syntax-ns#type_label'].lower()
        if (has_label or has_ns_label) and name in nobel_dict:
            nobel_sorted.append([nobel_dict[name]['year'], nobel_dict[name]['category'], birthYear, gender, name])

with open( 'Nobel_try_3.csv', 'w') as file:
    csvwriter = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_NONNUMERIC)
    csvwriter.writerow(['year', 'category', 'birthyear', 'gender','name'])
    for element in nobel_sorted: 
        csvwriter.writerow(element)


        
#year, category, birthyear, gender, name 
```
