```python
import argparse, json, logging, csv, re, sys, codecs

floatre = re.compile("^\d+\.\d+$")
intre = re.compile("^\d+$")

def read_header(file="h.txt"):
    header=[]
    for line in open(file):
        header.append(line.strip())
    logging.info("%d lines in header", len(header))
    return header

def process_csv(file, header):
    out=[]
    stdin = file == "-"
    fd = sys.stdin if stdin else codecs.open(file, 'r', 'UTF-8')
    reader = csv.reader(fd)
    for nr, row in enumerate(reader):
        logging.debug("%d fields in line %d", len(row), nr)
        d = dict()
        out.append(d)
        for i, field in enumerate(row):
            if field != "NULL":
                if floatre.match(field):
                    d[header[i]] = float(field)
                elif intre.match(field):
                    d[header[i]] = int(field)
                else:
                    d[header[i]] = field
    if not stdin:
        fd.close()
    return out

''' 
header = read_header('h.txt')
data = process_csv('years/1964', header)
for element in data:
    if 'award_label' in element:
        #print(element['award_label'])
        if "academy award" in element['award_label'].lower():
            print(element)
    if 'Crowe' in element['rdf-schema#label']:
-        print(element)
'''
```

```python
import json
import csv 

actors = {}
with open ('oscars/actors_names.csv') as file:
    csv_reader= csv.DictReader(file) 
    for actor in csv_reader:
        actor['Gender'] = 'M'
        del actor['Link']
        actors[actor['Actor']] = dict(actor)
with open ('oscars/actresses_names.csv') as file:
    csv_reader= csv.DictReader(file) 
    for actor in csv_reader:
        actor['Gender'] = 'F'
        del actor['Link']
        actors[actor['Actor']] = dict(actor)

header = read_header('h.txt')
actors_sorted= []
for year in range(1860, 2014):
    print(year)
    
    data = process_csv('years/' + str(year), header)
    for elements in data:
        name = elements['rdf-schema#label']
        birthYear = elements['birthDate'].split('-')[0][-4:]
        has_label = 'award_label' in elements and "academy award" in elements['award_label'].lower()
        has_ns_label = "actor" in elements['22-rdf-syntax-ns#type_label'].lower()
        if (has_label or has_ns_label) and name in actors:
            actors_sorted.append([actors[name]['Year'], birthYear, actors[name]['Gender'], name])

with open( 'Oscar_try_1.csv', 'w') as file:
    csvwriter = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_NONNUMERIC)
    csvwriter.writerow(['year', 'birthyear', 'gender','name'])
    for element in actors_sorted: 
        csvwriter.writerow(element)
```
